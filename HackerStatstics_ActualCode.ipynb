                  *** EMPIRE STATE BUILDING GAME or HACKER STATISTICS**

Imagine the following: **you're walking up** the empire state building to DataCamp HeadQuarters and you're playing a game with a friend.

You throw a dice **100 times**.
*   If it's 1 or 2 you'll go one step down.
*   If it's 3, 4, or 5, you'll go one step up.
*   If you throw a 6, you'll throw the die again and will walk up the resulting number of steps.
*   Of course, you **can't go lower than step number 0**. 
*   You're a bit clumsy and have a chance of **0.1% of falling down the stairs** when you make a move. **Falling down** means that you have to start again from **step 0**.
* Bet with your friend you'll reach **60 steps higher.** 

**Solution:**

 Simulation  : Run simulations and check at which fraction of simulation we can reach 60 steps. 

```
# 1. Generate random numbers while rolling dice. 
import numpy as np 
np.random.seed(123)
print(np.random.randint(1,7)) 

# randint is used to print an integer 
```

```
# 2. Connecting dice outcome with steps conditions or  [Determine your next move]

# Your next move depends on the nice on dice which you threw, 
# by this we can infer that number obtained on dice decided whether next move (or step) is possible of not.


import numpy as np 
step = 50   
dice = np.random.randint(0,7)

if dice <= 2: 
  step = step + 1
elif dice == 3 or dice == 4 or dice == 5 : 
  step = step - 1
else : 
  step = step + np.random.randint(1,7)

print(dice)
print(step)


# we need to assume we are at some point in the game so our step is at 50th position. 
# random generator decides if step needs to be taken or not. 
# control statements (if-else) provides conditions like take step forward(step + 1),  step backward (step + 1) etc.

```

```
# 3. RANDOM WALK

# Since we are dealing with simulation we need to automate this process.  All the obtained values need to be stored under single variable. 
# we need following: 
 # 1. Functionality which repeats process n times - a for loop which which runs in given range (range = 10 or 100 or 1000 )
 # 2. Obtained values can be storred in a list , 'outcomes in our case'.
          # Append functionality is used to add elements to the list. 

# Outcome of flipping a coin determines whether a step needs to be taken on not, since its always heads or tails, 
# lets assume heads -> ready to take step (so always random generator is assigned)
#             tails -> flip again 


import numpy as np 
np.random.seed(123)

outcomes = []
for x in range(10):
  coin = np.random.randint(0,2)
  if coin == 0 :
    outcomes.append("heads")
  else:
    outcomes.append("tails")
print(outcomes)    


# Output gives only a random values added to the list, 
# its not a random walk as the values are independednt of each other, which wont work out while rolling dice and taking steps. 
# we can deal with this by just taking the count of total trails generated while simulating the game. 

```

```
# 4.  Coversion of Random Steps into Random Walk (or) [RANDOM WALK]

# This step converts random steps into a random walk. 

import numpy as np

np.random.seed(123)

tails = [0] # at start of simulation the tails list is always empty. 

for x in range(10):
  coin = np.random.randint(0,2)
  print(coin)     #printing coins
  tails .append(tails[x] + coin)
print("\n") 
print(tails)  #printing tails


# QUESTION : What is the correlation between taking coin and updating tails list, there by relating to random walking. 
```

```
# 5. Simulating the Random Walk  or [THE NEXT STEP ]  

# walking always starts with step 0 
# simulation can be done using for loop over a range lets say 100. 

import numpy as np 
np.random.seed(123)

random_walk= [0] #an empty random_walk list

for x in range(100):             
  dice = np.random.randint(1,7)  #rolling the dice
  if dice <=2:                   #determning next step 
    step = step - 1
  elif dice <=5:
    step = step + 1
  else:
    step = step + np.random.randint(1,7)
    
  random_walk.append(step)   #appending random_walk list with step values  # written inside for loop

print(random_walk)

# this code gives the distance traversed after 100 dice throws 
```

```
# 6. Making sure you never go below zero   or [How low can you go?]

# We need to consider the case of not going below 0!, this can be done by using max().
 
import numpy as np 
np.random.seed(123)

random_walk = [0]

for x in range(100):
  step = random_walk[-1]
  dice = np.random.randint(1,7)

  if dice <= 2:
    step = max(0, step-1)   # Max() returns biggest among two arguments. 
  elif dice <=5:
    step = step + 1 
  else: 
    step = step + np.random.randint(1,7)

    random_walk.append(step)

print(random_walk)
```

```
# 8. Chances of reaching 60 steps high 

# Each random walk will end up on a random steps(or end-points ) , 
# when simulated 1000 times we end up with 1000 different end-points, 
# this distribution of final steps will help us achieve the goal of reaching 60 steps high.  


# 1000 steps is a pretty high range, a list that takes all the 1000 output values will surely be a larger dataset. 
#  inorder to complete the dataset we should include the obtained random_walk data points as well.
# to make it happen append random_walks to the bigger list (all_walks , in our case).

import numpy as np 
np.random.seed(123)


all_walks = []            # Initialize all_walks (don't change this line)

for i in range(10) :      # Simulate random walk 10 times

 # Previous Code   
    random_walk = [0]
    for x in range(100) :
        step = random_walk[-1]
        dice = np.random.randint(1,7)

        if dice <= 2:
            step = max(0, step - 1)
        elif dice <= 5:
            step = step + 1
        else:
            step = step + np.random.randint(1,7)
        random_walk.append(step)

    all_walks.append(random_walk)           # Append random_walk to all_walks

print(all_walks)                     # Print all_walks

 
```

```

# 9. Visualizing all_walks
# Now that we have list of lists as all_walks, upon converting it into NumPy array we can create interesting visualizations using MatPlotLib library. 

import matplotlib.pyplot as plt
import numpy as np 
np.random.seed(123)


all_walks = []            # Initialize all_walks (don't change this line)

for i in range(10) :      # Simulate random walk 10 times

 # Previous Code   
    random_walk = [0]
    for x in range(100) :
        step = random_walk[-1]
        dice = np.random.randint(1,7)

        if dice <= 2:
            step = max(0, step - 1)
        elif dice <= 5:
            step = step + 1
        else:
            step = step + np.random.randint(1,7)
        random_walk.append(step)

    all_walks.append(random_walk)           # Append random_walk to all_walks

print(all_walks)                     # Print all_walks
 
#VISUALIZATION PART: 


np_aw = np.array(all_walks)     #Convert all_walks to Numpy array: np_aw
plt.plot(np_aw)
plt.title("All_Walks")
plt.show()
plt.clf()

# Plot for Transpose np_aw   #Every row in transpose represents values after 10 random walks. 

np_aw_t = np.transpose(np_aw)
plt.plot(np_aw_t)
plt.title("All_Walks Transpose")
plt.show()
```

```
# 10.  Dealing with 0.01% chances of falling or [Implementing Chances]

# Since, there are 0.01% chances of falling down we need to check to make sure we shouldn't fall
# So we geneate a random numbers for the falling down scenario and if we get any such falling cases we will immeidately reset the value to step = 0 
# finally we append this set of falling steps to all_walks data-set. 
# This analysis is carried out to find the clumsiness points where falling is possible and avoid such scenarios. 

# IN THE CODE BELOW :

# 1. Since we're looking at clumsiness points we simulate random_walk many more times than before (250 times in our case , i.e range for for loop)
# 2. We generate random values betweeen 0 and 1 (since,  0.01 < 1) and check which data points are clumsy and failing,
#        i.   Above condition is enclosed in an if loop and STEP is initialized to zero in every failing case (step = 0 )
#        ii.  We then append steps to random walks. 
#        iii.  Then random walks is appended with(to) much bigger list ( i.e all_walks ) 

import matplotlib.pyplot as plt
import numpy as np 
np.random.seed(123)


all_walks = []            # Initialize all_walks (don't change this line)

for i in range(10) :      # Simulate random walk 10 times

    random_walk = [0]
    for x in range(250) :    # For loop range changed to 250 times from 100 times
        step = random_walk[-1]
        dice = np.random.randint(1,7)

        if dice <= 2:
            step = max(0, step - 1)
        elif dice <= 5:
            step = step + 1
        else:
            step = step + np.random.randint(1,7)
        random_walk.append(step)

        if np.random.rand() <= 0.01:      # IMPLEMENTING CLUMSINESS
          step = 0 
        random_walk.append(step) 

    all_walks.append(random_walk)           # Append random_walk to all_walks

print(all_walks)                     # Print all_walks
 

#VISUALIZATION PART: 

np_aw = np.array(all_walks)     #Convert all_walks to Numpy array: np_aw
plt.plot(np_aw)
plt.title("All_Walks")
plt.show()
plt.clf()

# Plot for Transpose np_aw

np_aw_t = np.transpose(np_aw)
plt.plot(np_aw_t)
plt.title("All_Walks Transpose")
plt.show()  

```

```
# 11. Odds of reaching 60 steps higher. 

# We can idenitfy the odds from the endpoints of all random walks which have been simulated. 

# Plotting a histogram with all the obtained endpoints helps us to study the distribution of all such odds. 

# To make our data more reliable we opt 500 simulations (i.e for loop range is 500 )

# In Code:
 
 # 1. The LAST ROW in Transpose of all_walks (last row in np_aw_t  array ) consists of end-points of all 500 simulation
 # 2. We simply select the last row from and convert it into a numpy array making them  convenient to plot histogram using all the end_points to study distributions.  



import matplotlib.pyplot as plt
import numpy as np 
np.random.seed(123)


all_walks = []            # Initialize all_walks (don't change this line)

for i in range(10) :      # Simulate random walk 10 times

    random_walk = [0]
    for x in range(500) :    # For loop range changed to 500 times from 250 times
        step = random_walk[-1]
        dice = np.random.randint(1,7)

        if dice <= 2:
            step = max(0, step - 1)
        elif dice <= 5:
            step = step + 1
        else:
            step = step + np.random.randint(1,7)
        random_walk.append(step)

        if np.random.rand() <= 0.01:      # IMPLEMENTING CLUMSINESS
          step = 0 
        random_walk.append(step) 

    all_walks.append(random_walk)           # Append random_walk to all_walks

 # print(all_walks)                     # Print all_walks


# ODDS OF REACHING 60 STEPS HIGHER

np_aw_t = np.transpose(np.array(all_walks))

# Select last row from np_aw_t: ends
ends = np.array(np_aw_t[-1,:])

# Plot histogram of ends, display plot
plt.title("Odds")
plt.hist(ends)
plt.show() 
```

```
# 12. Calculating  CHANCES OF ACHIEVING or [calculation of the odds]

# What are the chances that the obtained end-points obtained in above observatins are above 60 steps

#    1. End-points were obtained from the "ends" numpy array, from which histogram was plotted. 
#    2. To calculate the chances we need to find the number of integers which are greater than or equal to 60 , then divide them by 500. 
#          This basically equates to the MEAN

np.mean(ends > 60)  
```
